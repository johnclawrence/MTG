```python
#Structure Number, Name, Combo Piece, First Black Mana, Free Non Black Mana, bad black card for mox
deckList=[]
Agadeem=(4, "Agadeem's Awakening",0,1,0,1)
Spy=(4, "Balustrade Spy",1,0,0,1)
Bridge=(1, "Bridge from Below",0,0,0,0)
Crit=(3, "Cabal Ritual",0,0,0,1)
Therapy=(2, "Cabal Therapy",0,0,0,0)
Mox=(3, "Chrome Mox",0,1,1,0)
Rit=(4, "Dark Ritual",0,0,0,0)
Return=(1, "Dread Return",0,0,0,0)
ESG=(4, "Elvish Spirit Guide",0,0,1,0)
Petal=(4, "Lotus Petal",0,1,1,0)
Narc=(4, "Narcomoeba",0,0,0,0)
Npact=(4, "Pact of Negation",0,0,0,0)
SSG=(4, "Simian Spirit Guide",0,0,1,0)
Spact=(4, "Summoner's Pact",0,1,1,0)
Oracle=(1, "Thassa's Oracle",0,0,0,0)
TS=(4, "Thoughtseize",0,0,0,0)
Turntimber=(4, "Turntimber Symbiosis",0,0,1,0)
Informer=(4, "Undercity Informer",1,0,0,1)
Cantor=(1, "Wild Cantor",0,1,0,0)

deckList.append(Agadeem)
deckList.append(Spy)
deckList.append(Bridge)
deckList.append(Crit)
deckList.append(Therapy)
deckList.append(Mox)
deckList.append(Rit)
deckList.append(Return)
deckList.append(ESG)
deckList.append(Petal)
deckList.append(Narc)
deckList.append(Npact)
deckList.append(SSG)
deckList.append(Spact)
deckList.append(Oracle)
deckList.append(TS)
deckList.append(Turntimber)
deckList.append(Informer)
deckList.append(Cantor)
```

```python
def getHand(deck):
    hand=[]
    initialHand=[]

    for y in range(0,7):
        randomCard=random.choice(deck)
        hand.append(randomCard)
        initialHand.append(randomCard)
        deck.remove(randomCard)
    randomCard=random.choice(deck)
    draw1=randomCard
    deck.remove(randomCard)
    randomCard=random.choice(deck)
    draw2=randomCard
    deck.remove(randomCard)
    return (hand,initialHand,draw1,draw2)

def testEnabler(hand):
    hasEnabler=0
    removedCard=''
    for card in hand:
        hasEnabler=hasEnabler+card[2]
    if hasEnabler>0:
        try:
            hand.remove(Spy)
            removedCard=Spy
        except:
            hand.remove(Informer)
            removedCard=Informer
    return(hasEnabler,hand,removedCard)

def testBlackSource(hand):
    hasBlackSource=0
    blacksourceCards=''
    landDrop=0
    BlackMana=0
    for card in hand:
        hasBlackSource=hasBlackSource+card[3]
    if hasBlackSource>0:
        #There are a few paths for black mana.
        #Path 1 is Agadeem's Awakening
        if hand.count(Agadeem)>0:
            #print("Agadeem")
            hand.remove(Agadeem)
            blacksourceCards=[Agadeem]
            landDrop=1
            BlackMana=1
        #Path 2 is Lotus Petal
        if BlackMana==0 and hand.count(Petal)>0:
            hand.remove(Petal)
            blacksourceCards=[Petal]
            BlackMana=1
            #print("Lotus Petal")
        #Path 3 is Wild Cantor is in hand.
        if BlackMana==0 and hand.count(Cantor)==1:
            HasCantorMana=0
            for card in hand:
                HasCantorMana=HasCantorMana+card[4]
            if HasCantorMana>0:
                #So to get our first red/green mana at this point...
                #Path 3.1 Turntimber Symbiosis
                if hand.count(Turntimber)>0:
                    hand.remove(Turntimber)
                    hand.remove(Cantor)
                    blacksourceCards=[Cantor,Turntimber]
                    BlackMana=1
                    landDrop=1
                    #print("Turntimber",hand)
                #Path 3.2 SSG/ESG
                if hand.count(SSG)>0 and BlackMana==0:
                    hand.remove(SSG)
                    hand.remove(Cantor)
                    blacksourceCards=[Cantor,SSG]
                    BlackMana=1
                    #print("SSG",hand)
                if hand.count(ESG)>0 and BlackMana==0:
                    hand.remove(ESG)
                    hand.remove(Cantor)
                    blacksourceCards=[Cantor,ESG]
                    BlackMana=1
                    #print("ESG",hand)
                #Path 3.3 Summoner's Pact
                if hand.count(Spact)>0 and BlackMana==0:
                    hand.remove(Spact)
                    hand.remove(Cantor)
                    blacksourceCards=[Cantor,Spact]
                    BlackMana=1
                    #print("Pact",hand)
                #There is no Chrome Mox Path from here
                #Game 1 you will literally never be able to chrome mox for G/R
        #Path 4 is Chrome Moxing a bad card
        if hand.count(Mox)>0 and BlackMana==0:
            HasBadPitch=0
            for card in hand:
                HasBadPitch=HasBadPitch+card[5]
            if HasBadPitch>0:
                #So our bad cards in order of our desire to pitch them...
                #4.1 Spy
                if hand.count(Spy)>0 and BlackMana==0:
                    blacksourceCards=[Mox,Spy]
                    hand.remove(Spy)
                    hand.remove(Mox)
                    BlackMana=1
                #4.2 Informer
                if hand.count(Informer)>0 and BlackMana==0:
                    hand.remove(Informer)
                    hand.remove(Mox)
                    blacksourceCards=[Mox,Informer]
                    BlackMana=1

                #4.3 The The Second Thoughtseize
                if hand.count(TS)>1 and BlackMana==0:
                    hand.remove(TS)
                    hand.remove(Mox)
                    blacksourceCards=[Mox,TS]
                    BlackMana=1
                #4.4 Cabal Therapy
                if hand.count(Therapy)>0 and BlackMana==0:
                    hand.remove(Therapy)
                    hand.remove(Mox)
                    blacksourceCards=[Mox,Therapy]
                    BlackMana=1
                #4.5 Cabal Ritual
                if hand.count(Crit)>0 and BlackMana==0:
                    hand.remove(Crit)
                    hand.remove(Mox)
                    blacksourceCards=[Mox,Crit]
                    BlackMana=1
                if BlackMana==0:
                    print(hand)
        #Path 5 is Summoner's Pact
        if BlackMana==0 and hand.count(Spact)>0:
             #Path 5.1 Turntimber
             if hand.count(Turntimber)>0:
                    blacksourceCards=[Spact,Turntimber]
                    hand.remove(Turntimber)
                    hand.remove(Spact)
                    BlackMana=1
                    landDrop=1
                    #print("Turntimber",hand)
                #Path 5.2 SSG/ESG
             if hand.count(SSG)>0 and BlackMana==0:
                    blacksourceCards=[Spact,SSG]
                    hand.remove(SSG)
                    hand.remove(Spact)
                    BlackMana=1
                    #print("SSG",hand)
             if hand.count(ESG)>0 and BlackMana==0:
                    blacksourceCards=[Spact,ESG]
                    hand.remove(ESG)
                    hand.remove(Spact)
                    BlackMana=1
                    #print("ESG",hand)
                #Path 5.3 Summoner's Pact
             if hand.count(Spact)>1 and BlackMana==0:
                    blacksourceCards=[Spact,Spact]
                    hand.remove(Spact)
                    hand.remove(Spact)
                    BlackMana=1
        #Path 6 is Chrome Moxing a good card.
        if hand.count(Mox)>0 and BlackMana==0:
            #Path 6.x The First Thoughtseize
            if hand.count(TS)>0 and BlackMana==0:
                    blacksourceCards=[TS,Mox]
                    hand.remove(TS)
                    hand.remove(Mox)
                    BlackMana=1
            #Path 6.x Bridge
            if hand.count(TS)>0 and BlackMana==0:
                    hand.remove(Bridge)
                    hand.remove(Mox)
                    blacksourceCards=[Bridge,Mox]
                    BlackMana=1
            #Path 6.x Dark Ritual
            if hand.count(Rit)>0 and BlackMana==0:
                    hand.remove(Rit)
                    hand.remove(Mox)
                    BlackMana=1
                    blacksourceCards=[Rit,Mox]
    return(landDrop,BlackMana,hand,blacksourceCards)

def testManaSource(hand,landDrop,blackMana):
    manaSourceCards=[]
    Mana=0
    if blackMana>0:
        Mana=1
    #If you've got a land and can play it, play a land.
    if landDrop==0 and hand.count(Turntimber):
        hand.remove(Turntimber)
        manaSourceCards.append(Turntimber)
        Mana=Mana+1
    #If you've got a dark ritual, play a dark ritual
    while hand.count(Rit)>0 and Mana<4:
        hand.remove(Rit)
        manaSourceCards.append(Rit)
        Mana=Mana+2
    #If you've got SSG/ESG/Petal use them
    while hand.count(SSG)>0 and Mana<4:
        hand.remove(SSG)
        manaSourceCards.append(SSG)
        Mana=Mana+1
    while hand.count(ESG)>0 and Mana<4:
        hand.remove(ESG)
        manaSourceCards.append(ESG)
        Mana=Mana+1
    while hand.count(Petal)>0 and Mana<4:
        hand.remove(Petal)
        manaSourceCards.append(Petal)
        Mana=Mana+1
    #Next is Spact
    while hand.count(Spact)>0 and Mana<4:
        hand.remove(Spact)
        manaSourceCards.append(Spact)
        Mana=Mana+1
    #And Cabal ritual if Mana>1
    while hand.count(Crit)>0 and Mana<4 and Mana>1:
        hand.remove(Crit)
        manaSourceCards.append(Crit)
        Mana=Mana+1
    #Next is Chrome Mox with Dead Cards
    chromeLoop=0
    while hand.count(Mox)>0 and Mana<4 and chromeLoop<2:
        if hand.count(Mox)>0 and hand.count(Agadeem)>0:
            hand.remove(Mox)
            hand.remove(Agadeem)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Agadeem)
            Mana=Mana+1
        if hand.count(Mox)>0 and hand.count(Spy)>0:
            hand.remove(Mox)
            hand.remove(Spy)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Spy)
            Mana=Mana+1
        if hand.count(Mox)>0 and hand.count(Narc)>0:
            hand.remove(Mox)
            hand.remove(Narc)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Narc)
            Mana=Mana+1            
        if hand.count(Mox)>0 and hand.count(Informer)>0:
            hand.remove(Mox)
            hand.remove(Informer)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Informer)
            Mana=Mana+1
        if hand.count(Mox)>0 and hand.count(Turntimber)>0:
            hand.remove(Mox)
            hand.remove(Turntimber)
            Mana=Mana+1
            manaSourceCards.append(Mox)
            manaSourceCards.append(Turntimber)

    #Chrome Mox with Interaction
        if hand.count(Mox)>0 and hand.count(TS)>0:
            hand.remove(Mox)
            hand.remove(TS)
            manaSourceCards.append(Mox)
            manaSourceCards.append(TS)
            Mana=Mana+1 
        if hand.count(Mox)>0 and hand.count(Therapy)>0:
            hand.remove(Mox)
            hand.remove(Therapy)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Therapy)
            Mana=Mana+1   
        if hand.count(Mox)>0 and hand.count(Bridge)>0:
            hand.remove(Mox)
            hand.remove(Bridge)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Bridge)
            Mana=Mana+1
        if hand.count(Mox)>0 and hand.count(Npact)>0:
            hand.remove(Mox)
            hand.remove(Npact)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Npact)
            Mana=Mana+1     
    #Finally is Chrome Mox with Despiration
        if hand.count(Mox)>0 and hand.count(Cantor)>0 and Mana==3:
            hand.remove(Mox)
            hand.remove(Cantor)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Cantor)
            Mana=Mana+1
            
        if hand.count(Mox)>0 and hand.count(Crit)>0:
            hand.remove(Mox)
            hand.remove(Crit)
            manaSourceCards.append(Mox)
            manaSourceCards.append(Crit)
            Mana=Mana+1
        chromeLoop=chromeLoop+1
    #and one more stab Cabal ritual at the end.
    while hand.count(Crit)>0 and Mana<4 and Mana>1:
        hand.remove(Crit)
        Mana=Mana+1
    return(landDrop,Mana,hand,manaSourceCards)

def getGameParameters():
#Generate the Decklist
    fullDeckList=[]
    for cards in deckList:
        for eachCard in range(0,(cards[0])):
            fullDeckList.append(cards)
    #Draw your starting hand, and figure out what your first 2 draws would be.
    hand,initialHand,draw1,draw2=getHand(fullDeckList)
    #Do we have an enabler, and what is that enabler
    hasEnabler,hand,enablerCard=testEnabler(hand)
    #Do we have black mana, and what is the easiest way to get it.
    landDrop,blackMana,hand,blackSourceCards=testBlackSource(hand)
    #Can we get 3 generic mana from our hand and what is the easiest way to get it.
    landDrop,Mana,hand,manaSourceCards=testManaSource(hand,landDrop,blackMana)
    #If we can only get 2 generic mana can we win if we go to the next turn?
    Turn2Win=0
    if Mana==3 and (hand.count(Agadeem)>0 or hand.count(Turntimber)>0 or hand.count(Mox)>0 or hand.count(Cantor)>0):
            Turn2Win=1
    #Do we have protection?
    Protection=0
    if hand.count(TS)>0 or hand.count(Npact)>0:
        Protection=1
    #What is our final hand
    finalHand=[]
    for card in hand:
        finalHand.append(card[1])
    finalCount=len(finalHand)
    #So at this point, I have all of the key parameters of a hand.
        #hasEnabler is if we have the core combo piece
        #blackMana is if we have black mana
        #Mana is how much mana we have(with a limit of 4-5)
        #Turn2Win is if we have the Turn 2 Win.
        #finalCount is how many cards remain.
        #Is if we have protection
        #So... now I need to make decisions to get at what's the best way to play this deck
    return (hasEnabler,blackMana,Mana,Turn2Win,Protection,finalCount,draw1,draw2)
```
