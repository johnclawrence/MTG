```python
import random
```

```python
#Structure Number, Name, Combo Piece, First Black Mana, Free Non Black Mana, bad black card for mox
deckList=[]
Agadeem=(4, "Agadeem's Awakening",0,1,0,1)
Spy=(4, "Balustrade Spy",1,0,0,1)
Bridge=(1, "Bridge from Below",0,0,0,0)
Crit=(4, "Cabal Ritual",0,0,0,1)
Therapy=(2, "Cabal Therapy",0,0,0,0)
Mox=(2, "Chrome Mox",0,1,1,0)
Rit=(4, "Dark Ritual",0,0,0,0)
Return=(1, "Dread Return",0,0,0,0)
ESG=(4, "Elvish Spirit Guide",0,0,1,0)
Petal=(4, "Lotus Petal",0,1,1,0)
Narc=(4, "Narcomoeba",0,0,0,0)
Npact=(4, "Pact of Negation",0,0,0,0)
SSG=(4, "Simian Spirit Guide",0,0,1,0)
Spact=(4, "Summoner's Pact",0,1,1,0)
Oracle=(1, "Thassa's Oracle",0,0,0,0)
TS=(4, "Thoughtseize",0,0,0,0)
Turntimber=(4, "Turntimber Symbiosis",0,0,1,0)
Informer=(4, "Undercity Informer",1,0,0,1)
Cantor=(1, "Wild Cantor",0,1,0,0)

deckList.append(Agadeem)
deckList.append(Spy)
deckList.append(Bridge)
deckList.append(Crit)
deckList.append(Therapy)
deckList.append(Mox)
deckList.append(Rit)
deckList.append(Return)
deckList.append(ESG)
deckList.append(Petal)
deckList.append(Narc)
deckList.append(Npact)
deckList.append(SSG)
deckList.append(Spact)
deckList.append(Oracle)
deckList.append(TS)
deckList.append(Turntimber)
deckList.append(Informer)
deckList.append(Cantor)
```

```python
def getHand(deck):
    hand=[]
    initialHand=[]
    for y in range(0,7):
        randomCard=random.choice(deck)
        hand.append(randomCard)
        initialHand.append(randomCard)
        deck.remove(randomCard)
    return (hand,initialHand)

def testEnabler(hand):
    hasEnabler=0
    for card in hand:
        hasEnabler=hasEnabler+card[2]
    if hasEnabler>0:
        try:
            hand.remove(Spy)
        except:
            hand.remove(Informer)
    return(hasEnabler,hand)

def testBlackSource(hand,hasBlackSource):
    for card in hand:
        hasBlackSource=hasBlackSource+card[3]
    return (hasBlackSource)
```

```python
Loss=0
Mull7=0
Mull6=0
Mull5=0
Mull4=0
Mull3=0
t2Mull7=0
t2Mull6=0
t2Mull5=0
t2Mull4=0
t2Mull3=0
reportEvery=100
reportCount=0
fullDeckList=[]
handCount=100000


for x in range(0,handCount):
    noBlackSource=0
    blackSource=0
    fullDeckList=[]
    for cards in deckList:
        for eachCard in range(0,(cards[0])):
            fullDeckList.append(cards)
    reportCount=reportCount+1
    hand,initialHand=getHand(fullDeckList)
    
    hasBlackSource=0
    landDrop=0
    BlackMana=0
    Mana=0
    Turn2Win=0
    hasEnabler,hand=testEnabler(hand)
    
    if hasEnabler!=0:
        hasBlackSource=testBlackSource(hand,hasBlackSource)
        if hasBlackSource>0:
            #There are a few paths for black mana.
            #Path 1 is Agadeem's Awakening
            if hand.count(Agadeem)>0:
                #print("Agadeem")
                hand.remove(Agadeem)
                landDrop=1
                BlackMana=1
            #Path 2 is Lotus Petal
            if BlackMana==0 and hand.count(Petal)>0:
                hand.remove(Petal)
                BlackMana=1
                #print("Lotus Petal")
            #Path 3 is Wild Cantor is in hand.
            if BlackMana==0 and hand.count(Cantor)==1:
                HasCantorMana=0
                for card in hand:
                    HasCantorMana=HasCantorMana+card[4]
                if HasCantorMana>0:
                    #So to get our first red/green mana at this point...
                    #Path 3.1 Turntimber Symbiosis
                    if hand.count(Turntimber)>0:
                        hand.remove(Turntimber)
                        hand.remove(Cantor)
                        BlackMana=1
                        landDrop=1
                        #print("Turntimber",hand)
                    #Path 3.2 SSG/ESG
                    if hand.count(SSG)>0 and BlackMana==0:
                        hand.remove(SSG)
                        hand.remove(Cantor)
                        BlackMana=1
                        #print("SSG",hand)
                    if hand.count(ESG)>0 and BlackMana==0:
                        hand.remove(ESG)
                        hand.remove(Cantor)
                        BlackMana=1
                        #print("ESG",hand)
                    #Path 3.3 Summoner's Pact
                    if hand.count(Spact)>0 and BlackMana==0:
                        hand.remove(Spact)
                        hand.remove(Cantor)
                        BlackMana=1
                        #print("Pact",hand)
                    #There is no Chrome Mox Path from here
                    #Game 1 you will literally never be able to chrome mox for G/R
            #Path 4 is Chrome Moxing a bad card
            if hand.count(Mox)>0 and BlackMana==0:
                HasBadPitch=0
                for card in hand:
                    HasBadPitch=HasBadPitch+card[5]
                if HasBadPitch>0:
                    #So our bad cards in order of our desire to pitch them...
                    #4.1 Spy
                    if hand.count(Spy)>0 and BlackMana==0:
                        hand.remove(Spy)
                        hand.remove(Mox)
                        BlackMana=1
                    #4.2 Informer
                    if hand.count(Informer)>0 and BlackMana==0:
                        hand.remove(Informer)
                        hand.remove(Mox)
                        BlackMana=1
    
                    #4.3 The The Second Thoughtseize
                    if hand.count(TS)>1 and BlackMana==0:
                        hand.remove(TS)
                        hand.remove(Mox)
                        BlackMana=1
                    #4.4 Cabal Therapy
                    if hand.count(Therapy)>0 and BlackMana==0:
                        hand.remove(Therapy)
                        hand.remove(Mox)
                        BlackMana=1
                    #4.5 Cabal Ritual
                    if hand.count(Crit)>0 and BlackMana==0:
                        hand.remove(Crit)
                        hand.remove(Mox)
                        BlackMana=1
                    if BlackMana==0:
                        print(hand)
            #Path 5 is Summoner's Pact
            if BlackMana==0 and hand.count(Spact)>0:
                 #Path 5.1 Turntimber
                 if hand.count(Turntimber)>0:
                        hand.remove(Turntimber)
                        hand.remove(Spact)
                        BlackMana=1
                        landDrop=1
                        #print("Turntimber",hand)
                    #Path 5.2 SSG/ESG
                 if hand.count(SSG)>0 and BlackMana==0:
                        hand.remove(SSG)
                        hand.remove(Spact)
                        BlackMana=1
                        #print("SSG",hand)
                 if hand.count(ESG)>0 and BlackMana==0:
                        hand.remove(ESG)
                        hand.remove(Spact)
                        BlackMana=1
                        #print("ESG",hand)
                    #Path 5.3 Summoner's Pact
                 if hand.count(Spact)>1 and BlackMana==0:
                        hand.remove(Spact)
                        hand.remove(Spact)
                        BlackMana=1
            #Path 6 is Chrome Moxing a good card.
            if hand.count(Mox)>0 and BlackMana==0:
                #Path 6.x The First Thoughtseize
                if hand.count(TS)>0 and BlackMana==0:
                        hand.remove(TS)
                        hand.remove(Mox)
                        BlackMana=1
                #Path 6.x Bridge
                if hand.count(TS)>0 and BlackMana==0:
                        hand.remove(Bridge)
                        hand.remove(Mox)
                        BlackMana=1
                #Path 6.x Dark Ritual
                if hand.count(Rit)>0 and BlackMana==0:
                        hand.remove(Rit)
                        hand.remove(Mox)
                        BlackMana=1
            if BlackMana==0:
                noBlackSource=noBlackSource+1
    if BlackMana>0:
        Mana=1
        #If you've got a land, play a land.
        if landDrop==0 and hand.count(Turntimber):
            hand.remove(Turntimber)
            Mana=Mana+1
        #If you've got a dark ritual, play a dark ritual
        while hand.count(Rit)>0 and Mana<4:
            hand.remove(Rit)
            Mana=Mana+2
        #If you've got SSG/ESG/Petal use them
        while hand.count(SSG)>0 and Mana<4:
            hand.remove(SSG)
            Mana=Mana+1
        while hand.count(ESG)>0 and Mana<4:
            hand.remove(ESG)
            Mana=Mana+1
        while hand.count(Petal)>0 and Mana<4:
            hand.remove(Petal)
            Mana=Mana+1
        #Next is Spact
        while hand.count(Spact)>0 and Mana<4:
            hand.remove(Spact)
            Mana=Mana+1
        #And Cabal ritual if Mana>1
        while hand.count(Crit)>0 and Mana<4 and Mana>1:
            hand.remove(Crit)
            Mana=Mana+1
        #Next is Chrome Mox with Dead Cards
        chromeLoop=0
        while hand.count(Mox)>0 and Mana<4 and chromeLoop<2:
            if hand.count(Mox)>0 and hand.count(Agadeem)>0:
                hand.remove(Mox)
                hand.remove(Agadeem)
                Mana=Mana+1
            if hand.count(Mox)>0 and hand.count(Spy)>0:
                hand.remove(Mox)
                hand.remove(Spy)
                Mana=Mana+1
            if hand.count(Mox)>0 and hand.count(Cantor)>0:
                hand.remove(Mox)
                hand.remove(Cantor)
                Mana=Mana+1
            if hand.count(Mox)>0 and hand.count(Narc)>0:
                hand.remove(Mox)
                hand.remove(Narc)
                Mana=Mana+1            
            if hand.count(Mox)>0 and hand.count(Informer)>0:
                hand.remove(Mox)
                hand.remove(Informer)
                Mana=Mana+1
            if hand.count(Mox)>0 and hand.count(Turntimber)>0:
                hand.remove(Mox)
                hand.remove(Turntimber)
                Mana=Mana+1
  
        #Chrome Mox with Interaction
            if hand.count(Mox)>0 and hand.count(TS)>0:
                hand.remove(Mox)
                hand.remove(TS)
                Mana=Mana+1 
            if hand.count(Mox)>0 and hand.count(Therapy)>0:
                hand.remove(Mox)
                hand.remove(Therapy)
                Mana=Mana+1   
            if hand.count(Mox)>0 and hand.count(Bridge)>0:
                hand.remove(Mox)
                hand.remove(Bridge)
                Mana=Mana+1
            if hand.count(Mox)>0 and hand.count(Npact)>0:
                hand.remove(Mox)
                hand.remove(Npact)
                Mana=Mana+1     
        #Finally is Chrome Mox with Despiration
            if hand.count(Mox)>0 and hand.count(Crit)>0:
                hand.remove(Mox)
                hand.remove(Crit)
                Mana=Mana+1
            chromeLoop=chromeLoop+1
        #and one more stab Cabal ritual at the end.
        while hand.count(Crit)>0 and Mana<4 and Mana>1:
            hand.remove(Crit)
            Mana=Mana+1
        
        if Mana>=4:
            cleanfHand=[]
            for card in hand:
                cleanfHand.append(card[1])
        if Mana==3 and (hand.count(Agadeem)>0 or hand.count(Turntimber)>0 or hand.count(Mox)>0 or hand.count(Cantor)>0):
            Turn2Win=1
            
    #print(x,hasEnabler,BlackMana,Mana)        
    cleaniHand=[]
    for card in initialHand:
        cleaniHand.append(card[1])
    state="Failed"
    if hasEnabler==0:
        state=["Lose","No Enabler"]
    if BlackMana==0 and state=="Failed":
        state=["Lose","No Black Mana"]
    if Turn2Win==1 and state=="Failed":
        state=["Turn2Win",(len(cleanfHand)),cleanfHand]
    if Mana<4 and state=="Failed":
        state=["Lose",(len(cleanfHand)),cleanfHand]
    if Mana>=4 and state=="Failed":
        state=["Win",(len(cleanfHand)),cleanfHand]
    if state[0]=="Lose":
        Loss=Loss+1
    if state[0]=="Win":
        if state[1]==0:
            Mull7=Mull7+1
        if state[1]==1:
            Mull6=Mull6+1
        if state[1]==2:
            Mull5=Mull5+1
        if state[1]==3:
            Mull4=Mull4+1
        if state[1]==4:
            Mull3=Mull3+1
    if state[0]=="Turn2Win":
        if state[1]==0:
            t2Mull7=t2Mull7+1
        if state[1]==1:
            t2Mull6=t2Mull6+1
        if state[1]==2:
            t2Mull5=t2Mull5+1
        if state[1]==3:
            t2Mull4=t2Mull4+1
        if state[1]==4:
            t2Mull3=t2Mull3+1

    #print(state,cleaniHand)
#    if reportEvery==reportCount:
#        reportCount=0
#        print(x,Loss/x,Mull7/x,Mull6/x,Mull5/x,Mull4/x,Mull3/x)

    
print("Mulligan",Loss/handCount)
print("Win Rq7",Mull7/handCount)
print("Win Rq6",Mull6/handCount)
print("Win Rq5",Mull5/handCount)
print("Win Rq4",Mull4/handCount)
print("WinT2 Rq7",t2Mull7/handCount)
print("WinT2 Rq6",t2Mull6/handCount)
print("WinT2 Rq5",t2Mull5/handCount)
print("WinT2 Rq4",t2Mull4/handCount)
print("WinT2 Rq4",t2Mull3/handCount)
```
